```python
# Создание соединения c базой данных (ВНИМАНИЕ! Ваши данные для подключения с БД могут отличаться от данного примера)
import psycopg2 # драйвер для соединения с Postgresql
import getpass  # фича для маскирования пароля (использую для демонстрации в классе)

conn=psycopg2.connect(user='postgres',
                      database='test',
                      host='localhost',
                      port='5432',
                      password=getpass.getpass('Insert password: ')
                        )
print(conn.get_dsn_parameters())
```

```python
# Импортируем библиотеку pandas (работа с таблицами)
import pandas as pd

# Запрос кодов ценных бумаг (из таблицы order_log). Результат запроса представляет собой таблицу (DataFrame) 
query_sec="SELECT DISTINCT \"SECCODE\" FROM order_log ORDER BY \"SECCODE\" ASC"
sec_codes=pd.read_sql_query(query_sec,conn)

# Запрос заявок, поступивших до заданного времении по заданной бумаге (из таблицы order_log)  
query_orders="SELECT * FROM order_log WHERE \"SECCODE\"=\'GAZP\' AND \"TIME\"<100500000;"
orders=pd.read_sql_query(query_orders,conn)

# Обратите внимание, что специальные символы (например, кавычки) Python в текстовых переменных следует отделять символом \,
# чтобы те воспринимались как обычный текст.
```

```python
# Узнаем размерность ре
print(orders.shape)

# Смотрим на первые 10 строк  результатов зарпоса для снятых заявок ('ACTION'=0). 
orders[orders['ACTION']==0].head(10)
```

```python
# Построение книги лимитированных заявок (LOB - limit order book)

# Устанавливаем маску лимитных заявок заявках
no_mrk_ind=orders['PRICE']!=0
# Или так:
# orderd=orders[orders['PRICE']!=0]
# тогда no_mrk_ind далее нигде не нужен!

#  Таблица выставленных заявок
plc=orders[['ORDERNO','BUYSELL','PRICE','VOLUME']][(orders['ACTION']==1) & no_mrk_ind]

# Таблица отозванных заявок
wdr=orders[['ORDERNO','VOLUME']][(orders['ACTION']==0) & no_mrk_ind]

# Таблица исполненных объемов
trds=orders[['ORDERNO','VOLUME']][(orders['ACTION']==2) & no_mrk_ind] 
# учитываем, что одна и таже выставленная заявка может участвовать в разных сделках
trds=trds[['ORDERNO','VOLUME']].groupby(['ORDERNO']).sum()
```

```python
# Книга (aka стакан, aka LOB - Limit Order Book)
# Join таблиц
lob=pd.merge(plc,wdr, on='ORDERNO',how='left',suffixes=('_plc','_wdr'))
lob=pd.merge(lob,trds, on='ORDERNO', how='left')
lob=lob.rename(columns={'VOLUME':'VOLUME_trds'})
# Заменим NaN на 0
lob=lob.fillna(0)
lob['TOTAL_VOL']=lob['VOLUME_plc']-lob['VOLUME_wdr']-lob['VOLUME_trds']
# lob[lob['TOTAL_VOL']<0]
lob=lob[lob['TOTAL_VOL']>0]
# lob
lob=lob.pivot_table(index='PRICE',columns='BUYSELL',values='TOTAL_VOL',aggfunc='sum')
lob.sort_index(axis=0,inplace=True, ascending=False)
# lob
lob=lob.fillna(0)
lob
```

```python
# Найти лучшие цены на покупку и на продажу
# bid, ask = x['LOB'][x['LOB']['B']>0].index[0], x['LOB'][x['LOB']['S']>0].index[-1]
bid, ask = lob[lob['B']>0].index[0:5], lob[lob['S']>0].index[-6:-1]
[bid, ask]
lob.loc[ask.append(bid)]
```

```python
# В продолжение разговора: если в качестве аргумента функции задать открытое соединение с БД, то так тоже будет работать. 

# Список аргументов:
    # Код эмитента - sec
    # Время -time
    # Соединение с БД - conn

def fnc_bld_lob(sec,time,conn):
    import pandas as pd
    query="SELECT * FROM order_log WHERE \"TIME\"<="+str(time)+" AND \"SECCODE\"="+sec
    orders=pd.read_sql_query(query,conn)
    
    # Устанавливаем маску лимитных заявок заявках
    no_mrk_ind=orders['PRICE']!=0

    #  Таблица выставленных заявок
    plc=orders[['ORDERNO','BUYSELL','PRICE','VOLUME']][(orders['ACTION']==1) & no_mrk_ind]

    # Таблица отозванных заявок
    wdr=orders[['ORDERNO','VOLUME']][(orders['ACTION']==0) & no_mrk_ind]

    # Таблица исполненных объемов
    trds=orders[['ORDERNO','VOLUME']][(orders['ACTION']==2) & no_mrk_ind] 
    trds=trds[['ORDERNO','VOLUME']].groupby(['ORDERNO']).sum()
#     trds['ORDERNO']=trds.index
#     trds.index=range(trds.shape[0])

    # Книга (aka стакан, aka LOB)
    lob=pd.merge(plc,wdr, on='ORDERNO',how='left',suffixes=('_plc','_wdr'))
    lob=pd.merge(lob,trds, on='ORDERNO', how='left')
    lob=lob.rename(columns={'VOLUME':'VOLUME_trds'})
    lob=lob.fillna(0)
    lob['TOTAL_VOL']=lob['VOLUME_plc']-lob['VOLUME_wdr']-lob['VOLUME_trds']
    lob=lob[lob['TOTAL_VOL']>0]
    lob=lob.pivot_table(index='PRICE',columns='BUYSELL',values='TOTAL_VOL',aggfunc='sum')
    lob.sort_index(axis=0,inplace=True, ascending=False)
    lob=lob.fillna(0)
#     bid, ask = lob[lob['B']>0].index[0:5], lob[lob['S']>0].index[-6:-1]
#     [bid, ask]
#     lob=lob.loc[ask.append(bid)]
    
    results={'SECCODE': sec, 'TIME': time, 'QUERY': query, 'LOB': lob }
    return results
```

```python
# Проверяем работоспособность функции

login={'user':'postgres', 'database':'test', 'host':'localhost', 'port':'5432'}
rslt=fnc_bld_lob('\'GAZP\'',100500000,conn)
print(rslt)
print(rslt['QUERY'])
```
